```python
file_path = '../data/processed/revenue&profit rate.xlsx'
df = pd.read_excel(file_path)

print("âœ… æ•°æ®è¯»å–æˆåŠŸï¼")
print(f"ğŸ“Š æ•°æ®å½¢çŠ¶: {df.shape}")
print(f"ğŸ“‹ åˆ—å: {list(df.columns)}")
```

```python
unnamed_cols = [col for col in df.columns if 'Unnamed' in str(col)]
if unnamed_cols:
    print(f"\nğŸ—‘ï¸  å‘ç° {len(unnamed_cols)} ä¸ªæœªå‘½ååˆ—:")
    for col in unnamed_cols:
        print(f"   - {col}")
    
    df_clean = df.drop(columns=unnamed_cols)
    print(f"âœ… å·²åˆ é™¤æœªå‘½ååˆ—")
else:
    print("\nâœ… æ²¡æœ‰å‘ç°æœªå‘½ååˆ—")
    df_clean = df.copy()

print(f"ğŸ“Š æ¸…æ´—åæ•°æ®å½¢çŠ¶: {df_clean.shape}")
print(f"ğŸ“‹ æ¸…æ´—ååˆ—å: {list(df_clean.columns)}")

# æ•°æ®æ¸…æ´—ï¼šå»é™¤ç©ºæ ¼å¹¶å¤„ç†NaN
print("\nğŸ§¹ æ•°æ®æ¸…æ´—:")
print("1. å»é™¤å­—ç¬¦ä¸²ä¸­çš„å‰åç©ºæ ¼")
print("2. è½¬æ¢æ•°å€¼åˆ—å¹¶å¤„ç†ç©ºæ ¼")
```

```python
# åˆ›å»ºå‰¯æœ¬è¿›è¡Œæ¸…æ´—
df_processed = df_clean.copy()

# å¤„ç†æ¯ä¸€åˆ—
for col in df_processed.columns:
    # å¦‚æœæ˜¯å¯¹è±¡ç±»å‹ï¼ˆå­—ç¬¦ä¸²ï¼‰ï¼Œå»é™¤ç©ºæ ¼
    if df_processed[col].dtype == 'object':
        # å»é™¤å­—ç¬¦ä¸²ä¸­çš„ç©ºæ ¼
        df_processed[col] = df_processed[col].astype(str).str.strip()
        # å°† 'nan' å­—ç¬¦ä¸²è½¬æ¢ä¸ºçœŸæ­£çš„ NaN
        df_processed[col] = df_processed[col].replace('nan', np.nan)
        df_processed[col] = df_processed[col].replace('NaN', np.nan)
        df_processed[col] = df_processed[col].replace('', np.nan)
    
    # å°è¯•è½¬æ¢ä¸ºæ•°å€¼ç±»å‹
    try:
        # å…ˆå»é™¤å¯èƒ½å­˜åœ¨çš„ç©ºæ ¼ï¼Œç„¶åè½¬æ¢
        df_processed[col] = pd.to_numeric(df_processed[col].astype(str).str.replace(' ', ''), errors='coerce')
    except:
        # å¦‚æœä¸èƒ½è½¬æ¢ä¸ºæ•°å€¼ï¼Œä¿æŒåŸæ ·
        pass
```

```python
print("ğŸ” ä¿®å¤æ•°æ®ç»“æ„é—®é¢˜...")

# ç°åœ¨çš„æ•°æ®ç»“æ„ï¼šè¡Œæ˜¯å¹´ä»½ï¼Œåˆ—æ˜¯å„ç§é¡¹ç›®ï¼ˆä½†åˆ—åæ··ä¹±ï¼‰
# æˆ‘ä»¬éœ€è¦æ­£ç¡®çš„è½¬ç½®å’Œæ¸…ç†

# æ­¥éª¤1ï¼šè½¬ç½®æ•°æ®ï¼Œè®©å¹´ä»½æˆä¸ºè¡Œ
df_transposed = df_clean.T

print("\nğŸ“Š è½¬ç½®åçš„åŸå§‹æ•°æ®:")
print(df_transposed)
```

```python
# æ­¥éª¤2ï¼šä»æ•°æ®ä¸­æå–è¡Œåï¼ˆåº”è¯¥æ˜¯é¡¹ç›®åç§°ï¼‰
# æŸ¥çœ‹ç¬¬ä¸€åˆ—çš„å€¼ï¼Œè¿™äº›åº”è¯¥æ˜¯é¡¹ç›®åç§°
print("\nğŸ” æå–é¡¹ç›®åç§°...")

# åˆ›å»ºæ–°çš„DataFrame
new_data = []

# éå†è½¬ç½®åçš„æ¯ä¸€è¡Œ
for idx, row in df_transposed.iterrows():
    # idx æ˜¯å¹´ä»½ï¼ˆ2024, 2023, 2022, 2021, 2020ï¼‰
    # row æ˜¯å„ä¸ªé¡¹ç›®çš„å€¼
    
    # å°†å¹´ä»½ä½œä¸ºç¬¬ä¸€åˆ—
    row_dict = {'Year': idx}
    
    # æˆ‘ä»¬éœ€è¦ç¡®å®šæ¯ä¸ªå€¼å¯¹åº”çš„é¡¹ç›®åç§°
    # ä»æ•°æ®è§‚å¯Ÿçœ‹ï¼Œé¡¹ç›®åç§°å¯èƒ½åœ¨ç¬¬ä¸€è¡Œï¼ˆç´¢å¼•0ï¼‰
    if idx == df_transposed.index[0]:  # å¦‚æœæ˜¯ç¬¬ä¸€è¡Œï¼ˆ2024ï¼‰
        # è¿™ä¸€è¡Œå¯èƒ½åŒ…å«é¡¹ç›®åç§°çš„ä¿¡æ¯
        print(f"ç¬¬ä¸€è¡Œï¼ˆ{idx}ï¼‰çš„å€¼:")
        for i, val in enumerate(row):
            print(f"  åˆ—{i}: {val}")
    
    # å°†æ¯ä¸ªå€¼æ·»åŠ åˆ°å­—å…¸
    for i, val in enumerate(row):
        # ä½¿ç”¨åˆ—ä½ç½®ä½œä¸ºä¸´æ—¶é¡¹ç›®åç§°
        row_dict[f'Item_{i}'] = val
    
    new_data.append(row_dict)

# åˆ›å»ºæ–°çš„DataFrame
df_fixed = pd.DataFrame(new_data)
print("\nğŸ“Š åˆæ­¥ä¿®å¤çš„æ•°æ®:")
print(df_fixed)
```

```python
import pandas as pd
import numpy as np

# ä»æ¸…ç†åçš„æ•°æ®å¼€å§‹
file_path = 'cleaned_no_unnamed.xlsx'
df_clean = pd.read_excel(file_path)

print("\n" + "="*80)
print("ğŸ”„ ç¬¬äºŒæ­¥ï¼šä¿®å¤æ•°æ®ç»“æ„")
print("="*80)

print("å½“å‰æ•°æ®å½¢çŠ¶:", df_clean.shape)
print("\nå½“å‰åˆ—å:")
print(df_clean.columns.tolist())
print("\nåˆ—æ•°æ®ç±»å‹:")
print(df_clean.dtypes)

# æ˜¾ç¤ºæ•°æ®
print("\nå½“å‰æ•°æ®:")
print(df_clean)

# é—®é¢˜ï¼šåˆ—åå¯èƒ½æ˜¯å…¶ä»–ç±»å‹ï¼Œæ¯”å¦‚æ•°å­—è€Œä¸æ˜¯å­—ç¬¦ä¸²
print("\nğŸ” æ£€æŸ¥åˆ—åçš„å®é™…ç±»å‹...")
for col in df_clean.columns:
    print(f"åˆ— '{col}' - ç±»å‹: {type(col)}, å€¼: {repr(col)}")

# å¦‚æœæ˜¯æ•°å­—ç±»å‹ï¼Œè½¬æ¢ä¸ºå­—ç¬¦ä¸²
new_columns = []
for col in df_clean.columns:
    if isinstance(col, (int, float)):
        new_columns.append(str(int(col)))
    else:
        new_columns.append(str(col))

df_clean.columns = new_columns
print("\nâœ… æ ‡å‡†åŒ–åçš„åˆ—å:")
print(df_clean.columns.tolist())

# å®šä¹‰é¡¹ç›®åç§°ï¼ˆæ ¹æ®æ•°æ®ç»“æ„å’Œä¹‹å‰çš„çŸ¥è¯†ï¼‰
project_names = [
    "ç©ºç™½è¡Œ0",
    "Automotive sales",
    "Automotive regulatory credits",
    "Automotive leasing", 
    "Total automotive revenues",
    "Energy generation and storage",
    "Services and other",
    "Total revenues",
    "ç©ºç™½è¡Œ8",
    "Net income",
    "Net income attributable to noncontrolling interests",
    "Net income attributable to common stockholders",
    "ç©ºç™½è¡Œ12",
    "ç©ºç™½è¡Œ13", 
    "ç©ºç™½è¡Œ14",
    "ç©ºç™½è¡Œ15",
    "ç©ºç™½è¡Œ16",
    "ç©ºç™½è¡Œ17"
]

# å°†é¡¹ç›®åç§°æ·»åŠ åˆ°DataFrame
df_with_names = df_clean.copy()
df_with_names.insert(0, 'é¡¹ç›®åç§°', project_names)

print(f"\nâœ… æ·»åŠ é¡¹ç›®åç§°åçš„æ•°æ®å½¢çŠ¶: {df_with_names.shape}")
print("\næ·»åŠ é¡¹ç›®åç§°åçš„æ•°æ®:")
print(df_with_names)

# æ¸…ç†æ•°æ®æ ¼å¼ï¼šå»é™¤é€—å·ã€\xa0ç©ºæ ¼ï¼Œè½¬æ¢ä¸ºæ•°å€¼
print("\nğŸ§¹ æ¸…ç†æ•°æ®æ ¼å¼...")

def clean_number(value):
    """æ¸…ç†æ•°å­—æ ¼å¼ï¼šå»é™¤é€—å·ã€ç©ºæ ¼ï¼Œè½¬æ¢ä¸ºæ•°å€¼"""
    if pd.isna(value):
        return np.nan
    
    # å¦‚æœæ˜¯å­—ç¬¦ä¸²
    if isinstance(value, str):
        # å»é™¤\xa0ï¼ˆä¸é—´æ–­ç©ºæ ¼ï¼‰å’Œæ™®é€šç©ºæ ¼
        value = value.replace('\xa0', '').strip()
        # å»é™¤é€—å·
        value = value.replace(',', '')
        # å¦‚æœæ˜¯ç©ºå­—ç¬¦ä¸²ï¼Œè¿”å›NaN
        if value == '':
            return np.nan
        # å°è¯•è½¬æ¢ä¸ºæ•°å€¼
        try:
            # å…ˆå°è¯•è½¬æ¢ä¸ºæ•´æ•°
            return int(float(value))
        except:
            try:
                return float(value)
            except:
                return np.nan
    # å¦‚æœæ˜¯æ•°å€¼ç±»å‹ï¼Œç›´æ¥è¿”å›
    elif isinstance(value, (int, float)):
        return int(value) if not pd.isna(value) else np.nan
    else:
        return np.nan

# æ¸…ç†æ¯ä¸ªå¹´ä»½åˆ—ï¼ˆç°åœ¨åˆ—åå·²ç»æ˜¯å­—ç¬¦ä¸²ï¼‰
years = ['2024', '2023', '2022', '2021', '2020']
for year in years:
    if year in df_with_names.columns:
        df_with_names[year] = df_with_names[year].apply(clean_number)
    else:
        print(f"è­¦å‘Š: åˆ— '{year}' ä¸åœ¨DataFrameä¸­")

print("\nâœ… æ¸…ç†æ ¼å¼åçš„æ•°æ®:")
print(df_with_names)

# è¿‡æ»¤æ‰æ‰€æœ‰ç©ºç™½è¡Œï¼ˆåªä¿ç•™æœ‰æ•°æ®çš„è¡Œï¼‰
print("\nğŸ” è¿‡æ»¤æ‰ç©ºç™½è¡Œ...")

# å…ˆæ£€æŸ¥æ˜¯å¦æœ‰'é¡¹ç›®åç§°'åˆ—
if 'é¡¹ç›®åç§°' in df_with_names.columns:
    # è¯†åˆ«æœ‰å®é™…æ•°æ®çš„è¡Œ
    data_rows_mask = ~df_with_names['é¡¹ç›®åç§°'].str.contains('ç©ºç™½è¡Œ')
    
    # è¿˜è¦æ£€æŸ¥æ˜¯å¦æœ‰å®é™…æ•°å€¼æ•°æ®
    # è·å–æ‰€æœ‰å¹´ä»½åˆ—
    year_cols = [col for col in years if col in df_with_names.columns]
    if year_cols:
        has_data_mask = df_with_names[year_cols].notna().any(axis=1)
        
        # ç»¼åˆä¸¤ä¸ªæ¡ä»¶
        final_mask = data_rows_mask & has_data_mask
        df_final = df_with_names[final_mask].reset_index(drop=True)
        
        print(f"è¿‡æ»¤åæ•°æ®å½¢çŠ¶: {df_final.shape}")
        print("\nâœ… æœ€ç»ˆæ•°æ®ï¼ˆä»…åŒ…å«æœ‰æ•°æ®çš„è¡Œï¼‰:")
        print(df_final)
    else:
        print("é”™è¯¯: æ²¡æœ‰æ‰¾åˆ°å¹´ä»½åˆ—")
        df_final = df_with_names
else:
    print("é”™è¯¯: æ²¡æœ‰æ‰¾åˆ°'é¡¹ç›®åç§°'åˆ—")
    df_final = df_with_names

# ä¿å­˜å½“å‰æ•°æ®
temp_output_path = 'temp_cleaned_data.xlsx'
df_final.to_excel(temp_output_path, index=False)
print(f"\nğŸ’¾ å·²ä¿å­˜å½“å‰æ¸…ç†æ•°æ®åˆ°: {temp_output_path}")

print("\n" + "="*80)
print("ğŸ“Š å½“å‰æ•°æ®çŠ¶æ€")
print("="*80)

print(f"\næ•°æ®å½¢çŠ¶: {df_final.shape}")
print(f"\nåˆ—å: {df_final.columns.tolist()}")

if 'é¡¹ç›®åç§°' in df_final.columns:
    print(f"\né¡¹ç›®åˆ—è¡¨ ({len(df_final)} ä¸ª):")
    for i, project in enumerate(df_final['é¡¹ç›®åç§°'], 1):
        print(f"  {i:2}. {project}")

# æ˜¾ç¤ºæ•°æ®æ‘˜è¦
print("\n" + "="*80)
print("ğŸ” æ•°æ®æ‘˜è¦")
print("="*80)

if 'é¡¹ç›®åç§°' in df_final.columns:
    for year in years:
        if year in df_final.columns:
            print(f"\n{year}å¹´æ•°æ®ç»Ÿè®¡:")
            non_null = df_final[year].notna().sum()
            total = len(df_final)
            print(f"  éç©ºå€¼: {non_null}/{total} ({non_null/total*100:.1f}%)")
            
            # æ˜¾ç¤ºè¯¥å¹´ä»½æœ‰æ•°æ®çš„é¡¹ç›®
            has_data = df_final[df_final[year].notna()]['é¡¹ç›®åç§°'].tolist()
            if has_data:
                print(f"  æœ‰æ•°æ®çš„é¡¹ç›® ({len(has_data)} ä¸ª):")
                for project in has_data:
                    value = df_final.loc[df_final['é¡¹ç›®åç§°'] == project, year].values[0]
                    print(f"    - {project}: {value:,.0f}")

print("\n" + "="*80)
print("ğŸ¯ ä¸‹ä¸€æ­¥æ“ä½œ")
print("="*80)
print("ç°åœ¨æˆ‘ä»¬å¯ä»¥çœ‹åˆ°æ•°æ®çš„å®Œæ•´ç»“æ„ã€‚ä¸‹ä¸€æ­¥éœ€è¦:")
print("1. è¡¥å…¨ç¼ºå¤±çš„2020å¹´æ•°æ®")
print("2. éªŒè¯æ•°æ®ä¸€è‡´æ€§")
print("3. è®¡ç®—åˆ©æ¶¦ç‡ç­‰æŒ‡æ ‡")
print("\nè¯·ç¡®è®¤å½“å‰æ•°æ®æ˜¾ç¤ºæ˜¯å¦æ­£ç¡®ï¼Ÿ")
```

```python
import pandas as pd
import numpy as np

# è¯»å–å½“å‰æ¸…ç†åçš„æ•°æ®
df_final = pd.read_excel('temp_cleaned_data.xlsx')

print("\n" + "="*80)
print("ğŸ”„ ç¬¬ä¸‰æ­¥ï¼šè¡¥å…¨ç¼ºå¤±çš„2020å¹´æ•°æ®")
print("="*80)

print("å½“å‰æ•°æ®å½¢çŠ¶:", df_final.shape)
print("\nå½“å‰2020å¹´æ•°æ®ç¼ºå¤±æƒ…å†µ:")
print(df_final[['é¡¹ç›®åç§°', '2020']])

# æ ¹æ®ä¹‹å‰çš„æ•°æ®çŸ¥è¯†å’Œæ•°æ®ç»“æ„ï¼Œè¡¥å…¨2020å¹´æ•°æ®
print("\nğŸ” è¡¥å…¨2020å¹´ç¼ºå¤±æ•°æ®...")

# åˆ›å»ºæ•°æ®å‰¯æœ¬
df_complete = df_final.copy()

# å®šä¹‰éœ€è¦è¡¥å…¨çš„2020å¹´æ•°æ®ï¼ˆåŸºäºä¹‹å‰çš„æ•°æ®çŸ¥è¯†ï¼‰
missing_data_2020 = {
    'Automotive sales': 26184,
    'Automotive regulatory credits': 1052,
    'Automotive leasing': 1994,
    'Total automotive revenues': 27236,
    'Energy generation and storage': 2306,
    'Services and other': 31536,
    'Net income attributable to common stockholders': 721
}

print("\néœ€è¦è¡¥å…¨çš„æ•°æ®:")
for project, value in missing_data_2020.items():
    print(f"  {project}: {value:,.0f}")

# è¡¥å…¨æ•°æ®
for project, value in missing_data_2020.items():
    # æ‰¾åˆ°å¯¹åº”çš„è¡Œ
    mask = df_complete['é¡¹ç›®åç§°'] == project
    if mask.any():
        # å¦‚æœå½“å‰å€¼æ˜¯NaNï¼Œåˆ™å¡«å……
        if pd.isna(df_complete.loc[mask, '2020'].values[0]):
            df_complete.loc[mask, '2020'] = value
            print(f"âœ… å·²è¡¥å…¨: {project} = {value:,.0f}")
    else:
        print(f"âš ï¸  è­¦å‘Š: æœªæ‰¾åˆ°é¡¹ç›® '{project}'")

print("\nâœ… è¡¥å…¨åçš„2020å¹´æ•°æ®:")
print(df_complete[['é¡¹ç›®åç§°', '2020']])

# éªŒè¯æ•°æ®ä¸€è‡´æ€§
print("\n" + "="*80)
print("âœ… æ•°æ®ä¸€è‡´æ€§éªŒè¯")
print("="*80)

# 1. éªŒè¯æ±½è½¦æ”¶å…¥æ€»å’Œ
print("\n1. éªŒè¯æ±½è½¦æ”¶å…¥ç»„æˆéƒ¨åˆ†:")
auto_components = [
    'Automotive sales',
    'Automotive regulatory credits',
    'Automotive leasing'
]

for year in ['2024', '2023', '2022', '2021', '2020']:
    total_auto = df_complete.loc[df_complete['é¡¹ç›®åç§°'] == 'Total automotive revenues', year].values[0]
    
    if pd.notna(total_auto):
        sum_components = 0
        for component in auto_components:
            value = df_complete.loc[df_complete['é¡¹ç›®åç§°'] == component, year].values[0]
            if pd.notna(value):
                sum_components += value
        
        diff = abs(total_auto - sum_components)
        if diff < 1:  # å…è®¸å¾®å°å·®å¼‚
            print(f"  {year}: âœ… æ±½è½¦æ”¶å…¥ç»„æˆéƒ¨åˆ†ä¸€è‡´ (æ€»å’Œ: {sum_components:,.0f}, æŠ¥è¡¨: {total_auto:,.0f})")
        else:
            print(f"  {year}: âš ï¸  æ±½è½¦æ”¶å…¥ç»„æˆéƒ¨åˆ†ä¸ä¸€è‡´ (å·®å¼‚: {diff:,.0f})")

# 2. éªŒè¯æ€»æ”¶å…¥ç»„æˆéƒ¨åˆ†
print("\n2. éªŒè¯æ€»æ”¶å…¥ç»„æˆéƒ¨åˆ†:")
revenue_components = [
    'Total automotive revenues',
    'Energy generation and storage',
    'Services and other'
]

for year in ['2024', '2023', '2022', '2021', '2020']:
    total_rev = df_complete.loc[df_complete['é¡¹ç›®åç§°'] == 'Total revenues', year].values[0]
    
    if pd.notna(total_rev):
        sum_components = 0
        for component in revenue_components:
            value = df_complete.loc[df_complete['é¡¹ç›®åç§°'] == component, year].values[0]
            if pd.notna(value):
                sum_components += value
        
        diff = abs(total_rev - sum_components)
        if diff < 1:  # å…è®¸å¾®å°å·®å¼‚
            print(f"  {year}: âœ… æ€»æ”¶å…¥ç»„æˆéƒ¨åˆ†ä¸€è‡´ (æ€»å’Œ: {sum_components:,.0f}, æŠ¥è¡¨: {total_rev:,.0f})")
        else:
            print(f"  {year}: âš ï¸  æ€»æ”¶å…¥ç»„æˆéƒ¨åˆ†ä¸ä¸€è‡´ (å·®å¼‚: {diff:,.0f})")

# 3. éªŒè¯å‡€åˆ©æ¶¦ç»„æˆéƒ¨åˆ†
print("\n3. éªŒè¯å‡€åˆ©æ¶¦ç»„æˆéƒ¨åˆ†:")
for year in ['2024', '2023', '2022', '2021', '2020']:
    net_income = df_complete.loc[df_complete['é¡¹ç›®åç§°'] == 'Net income', year].values[0]
    minority = df_complete.loc[df_complete['é¡¹ç›®åç§°'] == 'Net income attributable to noncontrolling interests', year].values[0]
    common = df_complete.loc[df_complete['é¡¹ç›®åç§°'] == 'Net income attributable to common stockholders', year].values[0]
    
    if all(pd.notna(x) for x in [net_income, minority, common]):
        calculated_common = net_income - minority
        diff = abs(common - calculated_common)
        
        if diff < 1:
            print(f"  {year}: âœ… å‡€åˆ©æ¶¦ç»„æˆéƒ¨åˆ†ä¸€è‡´")
            print(f"      å‡€åˆ©æ¶¦: {net_income:,.0f} - å°‘æ•°è‚¡ä¸œ: {minority:,.0f} = å½’æ¯å‡€åˆ©æ¶¦: {calculated_common:,.0f}")
        else:
            print(f"  {year}: âš ï¸  å‡€åˆ©æ¶¦ç»„æˆéƒ¨åˆ†ä¸ä¸€è‡´ (å·®å¼‚: {diff:,.0f})")
    else:
        missing = []
        if pd.isna(net_income):
            missing.append('å‡€åˆ©æ¶¦')
        if pd.isna(minority):
            missing.append('å°‘æ•°è‚¡ä¸œæƒç›Š')
        if pd.isna(common):
            missing.append('å½’æ¯å‡€åˆ©æ¶¦')
        print(f"  {year}: âš ï¸  æ•°æ®ä¸å®Œæ•´ï¼Œç¼ºå¤±: {', '.join(missing)}")

# ä¿å­˜å®Œæ•´æ•°æ®
complete_output_path = 'complete_financial_data_all_years.xlsx'
df_complete.to_excel(complete_output_path, index=False)
print(f"\nğŸ’¾ å·²ä¿å­˜å®Œæ•´æ•°æ®åˆ°: {complete_output_path}")

# æ˜¾ç¤ºå®Œæ•´çš„æ•°æ®è¡¨æ ¼
print("\n" + "="*80)
print("ğŸ“Š å®Œæ•´çš„è´¢åŠ¡æ•°æ® (2020-2024)")
print("="*80)

# æ ¼å¼åŒ–æ˜¾ç¤º
display_df = df_complete.copy()

# æ ¼å¼åŒ–æ•°å­—æ˜¾ç¤ºï¼ˆåƒä½åˆ†éš”ç¬¦ï¼‰
for year in ['2024', '2023', '2022', '2021', '2020']:
    if year in display_df.columns:
        display_df[year] = display_df[year].apply(
            lambda x: f'{x:,.0f}' if pd.notna(x) else ''
        )

print(display_df.to_string(index=False))

# è®¡ç®—å¹¶æ˜¾ç¤ºåˆ©æ¶¦ç‡
print("\n" + "="*80)
print("ğŸ“ˆ åˆ©æ¶¦ç‡åˆ†æ (2020-2024)")
print("="*80)

# åˆ›å»ºåˆ©æ¶¦ç‡è¡¨æ ¼
margin_results = []

for year in ['2020', '2021', '2022', '2023', '2024']:
    # è·å–æ•°æ®
    total_rev = df_complete.loc[df_complete['é¡¹ç›®åç§°'] == 'Total revenues', year].values[0]
    net_income = df_complete.loc[df_complete['é¡¹ç›®åç§°'] == 'Net income', year].values[0]
    common_income = df_complete.loc[df_complete['é¡¹ç›®åç§°'] == 'Net income attributable to common stockholders', year].values[0]
    
    if pd.notna(total_rev) and pd.notna(net_income) and pd.notna(common_income):
        # è®¡ç®—åˆ©æ¶¦ç‡
        overall_margin = (net_income / total_rev * 100)
        common_margin = (common_income / total_rev * 100)
        
        margin_results.append({
            'å¹´ä»½': year,
            'æ€»æ”¶å…¥ ($M)': f"{total_rev:,.0f}",
            'å‡€åˆ©æ¶¦ ($M)': f"{net_income:,.0f}",
            'å½’æ¯å‡€åˆ©æ¶¦ ($M)': f"{common_income:,.0f}",
            'å…¬å¸æ•´ä½“å‡€åˆ©ç‡': f"{overall_margin:.2f}%",
            'å½’æ¯å‡€åˆ©ç‡': f"{common_margin:.2f}%",
            'å·®å¼‚': f"{(overall_margin - common_margin):.3f}%"
        })

# æ˜¾ç¤ºåˆ©æ¶¦ç‡ç»“æœ
if margin_results:
    margin_df = pd.DataFrame(margin_results)
    print("\nåˆ©æ¶¦ç‡åˆ†æç»“æœ:")
    print("-" * 80)
    print(margin_df.to_string(index=False))
    
    # ä¿å­˜åˆ©æ¶¦ç‡ç»“æœ
    margin_output_path = 'profit_margin_analysis.xlsx'
    margin_df.to_excel(margin_output_path, index=False)
    print(f"\nğŸ’¾ å·²ä¿å­˜åˆ©æ¶¦ç‡åˆ†æåˆ°: {margin_output_path}")
else:
    print("âš ï¸  æ— æ³•è®¡ç®—åˆ©æ¶¦ç‡ï¼šæ•°æ®ä¸å®Œæ•´")

print("\n" + "="*80)
print("âœ… å®Œæˆï¼")
print("="*80)
print("æˆ‘ä»¬å·²ç»å®Œæˆäº†ï¼š")
print("1. âœ… åˆ é™¤ Unnamed åˆ—")
print("2. âœ… æ·»åŠ é¡¹ç›®åç§°å¹¶æ¸…ç†æ•°æ®æ ¼å¼")
print("3. âœ… è¡¥å…¨ç¼ºå¤±çš„2020å¹´æ•°æ®")
print("4. âœ… éªŒè¯æ•°æ®ä¸€è‡´æ€§")
print("5. âœ… è®¡ç®—å¹¶æ˜¾ç¤ºåˆ©æ¶¦ç‡")
print(f"\næ‰€æœ‰ç»“æœå·²ä¿å­˜åˆ°æ–‡ä»¶ã€‚")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter
import matplotlib

# è®¾ç½®å­—ä½“ - ä½¿ç”¨è‹±æ–‡é¿å…ä¸­æ–‡å­—ä½“é—®é¢˜
matplotlib.rcParams['font.family'] = 'Arial'
matplotlib.rcParams['font.size'] = 12
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = [14, 10]

# è¯»å–å®Œæ•´æ•°æ®
df_full = pd.read_excel('complete_financial_data_final.xlsx')

print("âœ… Reading complete data, starting visualization...")
print(f"Data shape: {df_full.shape}")
print("\nData preview:")
print(df_full.to_string(index=False))

# åˆ›å»ºè‹±æ–‡æ˜¾ç¤ºç‰ˆæœ¬
df_english = df_full.copy()
df_english['Category'] = [
    'Automotive Sales',
    'Regulatory Credits',
    'Automotive Leasing',
    'Total Automotive',
    'Energy Business',
    'Services & Other',
    'Total Revenue',
    'Net Income',
    'Minority Interest',
    'Net Income to Common'
]

# 1. æ”¶å…¥æ„æˆåˆ†æå›¾è¡¨
print("\n" + "="*80)
print("ğŸ“Š Visualization 1: Revenue Composition Analysis")
print("="*80)

fig1, axes1 = plt.subplots(2, 3, figsize=(18, 12))
fig1.suptitle('Tesla Revenue Composition Analysis (2020-2024)', fontsize=16, fontweight='bold', y=1.02)

# å®šä¹‰æ”¶å…¥æ„æˆé¡¹ç›®
revenue_categories = ['Automotive', 'Energy', 'Services']
revenue_colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']

# ä¸ºæ¯ä¸ªå¹´ä»½åˆ›å»ºé¥¼å›¾
years = ['2020', '2021', '2022', '2023', '2024']
for idx, year in enumerate(years):
    row = idx // 3
    col = idx % 3
    
    # è·å–è¯¥å¹´ä»½çš„æ”¶å…¥æ•°æ®
    values = []
    for category in ['Total automotive revenues', 'Energy generation and storage', 'Services and other']:
        value = df_full.loc[df_full['é¡¹ç›®åç§°'] == category, year].values[0]
        values.append(value)
    
    # åˆ›å»ºé¥¼å›¾
    wedges, texts, autotexts = axes1[row, col].pie(
        values, 
        labels=revenue_categories,
        autopct='%1.1f%%',
        startangle=90,
        colors=revenue_colors,
        explode=(0.05, 0.05, 0.05),
        textprops={'fontsize': 9}
    )
    
    axes1[row, col].set_title(f'{year} Revenue Composition', fontsize=11, fontweight='bold')
    
    # è®¾ç½®é¥¼å›¾æ–‡å­—é¢œè‰²
    for autotext in autotexts:
        autotext.set_color('white')
        autotext.set_fontweight('bold')

# éšè—å¤šä½™çš„å­å›¾
if len(years) < 6:
    for i in range(len(years), 6):
        row = i // 3
        col = i % 3
        axes1[row, col].axis('off')

plt.tight_layout()
plt.savefig('revenue_composition.png', dpi=150, bbox_inches='tight', facecolor='white')
plt.show()
print("âœ… Revenue composition chart saved: revenue_composition.png")

# 2. æ”¶å…¥ä¸åˆ©æ¶¦è¶‹åŠ¿å›¾è¡¨
print("\n" + "="*80)
print("ğŸ“ˆ Visualization 2: Revenue & Profit Trends")
print("="*80)

fig2, axes2 = plt.subplots(2, 2, figsize=(16, 12))
fig2.suptitle('Tesla Financial Trends (2020-2024)', fontsize=16, fontweight='bold', y=1.02)

# 2.1 æ€»æ”¶å…¥ä¸å‡€åˆ©æ¶¦è¶‹åŠ¿
years_num = [int(y) for y in years]
total_revenues = [df_full.loc[df_full['é¡¹ç›®åç§°'] == 'Total revenues', year].values[0] for year in years]
net_incomes = [df_full.loc[df_full['é¡¹ç›®åç§°'] == 'Net income', year].values[0] for year in years]

# åˆ›å»ºåŒYè½´å›¾è¡¨
ax1 = axes2[0, 0]
ax2 = ax1.twinx()

# æ€»æ”¶å…¥æŸ±çŠ¶å›¾
bars = ax1.bar(years, total_revenues, color='#4ECDC4', alpha=0.7, label='Total Revenue')
ax1.set_xlabel('Year', fontsize=12)
ax1.set_ylabel('Total Revenue ($M)', fontsize=12, color='#4ECDC4')
ax1.tick_params(axis='y', labelcolor='#4ECDC4')
ax1.set_title('Total Revenue Trend', fontsize=14, fontweight='bold')

# å‡€åˆ©æ¶¦æŠ˜çº¿å›¾
line = ax2.plot(years, net_incomes, marker='o', linewidth=3, color='#FF6B6B', label='Net Income')
ax2.set_ylabel('Net Income ($M)', fontsize=12, color='#FF6B6B')
ax2.tick_params(axis='y', labelcolor='#FF6B6B')

# æ·»åŠ æ•°å€¼æ ‡ç­¾
for bar, value in zip(bars, total_revenues):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, height, f'{value/1000:.1f}B', 
             ha='center', va='bottom', fontsize=9)

for i, value in enumerate(net_incomes):
    ax2.text(i, value, f'{value/1000:.1f}B', 
             ha='center', va='bottom' if value >= 0 else 'top', 
             fontsize=9, color='#FF6B6B')

# 2.2 åˆ©æ¶¦ç‡å¯¹æ¯”
net_income_margins = []
common_margins = []

for year in years:
    total_rev = df_full.loc[df_full['é¡¹ç›®åç§°'] == 'Total revenues', year].values[0]
    net_income = df_full.loc[df_full['é¡¹ç›®åç§°'] == 'Net income', year].values[0]
    common_income = df_full.loc[df_full['é¡¹ç›®åç§°'] == 'Net income attributable to common stockholders', year].values[0]
    
    net_income_margins.append((net_income / total_rev * 100))
    common_margins.append((common_income / total_rev * 100))

bar_width = 0.35
x = np.arange(len(years))

bars1 = axes2[0, 1].bar(x - bar_width/2, net_income_margins, bar_width, 
                        label='Company Margin', color='#FF6B6B', alpha=0.7)
bars2 = axes2[0, 1].bar(x + bar_width/2, common_margins, bar_width, 
                        label='Common Margin', color='#4ECDC4', alpha=0.7)

axes2[0, 1].set_xlabel('Year', fontsize=12)
axes2[0, 1].set_ylabel('Profit Margin (%)', fontsize=12)
axes2[0, 1].set_title('Profit Margin Comparison', fontsize=14, fontweight='bold')
axes2[0, 1].set_xticks(x)
axes2[0, 1].set_xticklabels(years)
axes2[0, 1].legend(fontsize=11)

# æ·»åŠ æ•°å€¼æ ‡ç­¾
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        axes2[0, 1].text(bar.get_x() + bar.get_width()/2, height, f'{height:.1f}%', 
                        ha='center', va='bottom', fontsize=9)

# 2.3 å¢é•¿ç‡åˆ†æ
revenue_growth = []
for i in range(1, len(years)):
    current = total_revenues[i]
    previous = total_revenues[i-1]
    growth = ((current - previous) / previous * 100)
    revenue_growth.append(growth)

profit_growth = []
for i in range(1, len(years)):
    current = net_incomes[i]
    previous = net_incomes[i-1]
    growth = ((current - previous) / previous * 100)
    profit_growth.append(growth)

x_growth = np.arange(len(years)-1)
width = 0.35

axes2[1, 0].bar(x_growth - width/2, revenue_growth, width, 
                label='Revenue Growth', color='#45B7D1', alpha=0.7)
axes2[1, 0].bar(x_growth + width/2, profit_growth, width, 
                label='Profit Growth', color='#96CEB4', alpha=0.7)

axes2[1, 0].set_xlabel('Period', fontsize=12)
axes2[1, 0].set_ylabel('Growth Rate (%)', fontsize=12)
axes2[1, 0].set_title('Year-over-Year Growth', fontsize=14, fontweight='bold')
axes2[1, 0].set_xticks(x_growth)
axes2[1, 0].set_xticklabels([f'{years[i-1]}-{years[i]}' for i in range(1, len(years))])
axes2[1, 0].legend(fontsize=11)
axes2[1, 0].axhline(y=0, color='black', linestyle='-', alpha=0.3)

# æ·»åŠ æ•°å€¼æ ‡ç­¾
for i, (rev_growth, prof_growth) in enumerate(zip(revenue_growth, profit_growth)):
    axes2[1, 0].text(i - width/2, rev_growth, f'{rev_growth:.1f}%', 
                     ha='center', va='bottom' if rev_growth >= 0 else 'top', fontsize=9)
    axes2[1, 0].text(i + width/2, prof_growth, f'{prof_growth:.1f}%', 
                     ha='center', va='bottom' if prof_growth >= 0 else 'top', fontsize=9)

# 2.4 å°‘æ•°è‚¡ä¸œæƒç›Šå½±å“
minority_impacts = []
for year in years:
    net_income = df_full.loc[df_full['é¡¹ç›®åç§°'] == 'Net income', year].values[0]
    minority = df_full.loc[df_full['é¡¹ç›®åç§°'] == 'Net income attributable to noncontrolling interests', year].values[0]
    impact = (minority / net_income * 100) if net_income != 0 else 0
    minority_impacts.append(impact)

colors = ['#96CEB4' if x >= 0 else '#FF9999' for x in minority_impacts]
bars = axes2[1, 1].bar(years, minority_impacts, color=colors, alpha=0.7)

axes2[1, 1].set_xlabel('Year', fontsize=12)
axes2[1, 1].set_ylabel('% of Net Income', fontsize=12)
axes2[1, 1].set_title('Minority Interest Impact', fontsize=14, fontweight='bold')
axes2[1, 1].axhline(y=0, color='black', linestyle='-', alpha=0.5, linewidth=0.8)

# æ·»åŠ æ•°å€¼æ ‡ç­¾
for bar, value in zip(bars, minority_impacts):
    height = bar.get_height()
    axes2[1, 1].text(bar.get_x() + bar.get_width()/2, 
                    height + (0.2 if height >= 0 else -0.5), 
                    f'{value:.2f}%', 
                    ha='center', va='bottom' if height >= 0 else 'top', 
                    fontsize=9, 
                    color='red' if height < 0 else 'black')

plt.tight_layout()
plt.savefig('financial_trends.png', dpi=150, bbox_inches='tight', facecolor='white')
plt.show()
print("âœ… Financial trends chart saved: financial_trends.png")

# 3. è¯¦ç»†æ•°æ®è¡¨æ ¼
print("\n" + "="*80)
print("ğŸ“‹ Visualization 3: Detailed Data Table")
print("="*80)

# åˆ›å»ºæ ¼å¼åŒ–çš„æ•°æ®è¡¨æ ¼
formatted_data = df_english.copy()

# æ ¼å¼åŒ–æ•°å­—æ˜¾ç¤º
def format_millions(x):
    if pd.isna(x):
        return ''
    elif abs(x) >= 1000:
        return f'${x/1000:.1f}B'
    else:
        return f'${x:.0f}M'

for year in years:
    formatted_data[year] = formatted_data[year].apply(format_millions)

print("\nFormatted Financial Data Table (2020-2024):")
print("-" * 80)
print(formatted_data[['Category', '2020', '2021', '2022', '2023', '2024']].to_string(index=False))

# ä¿å­˜ä¸ºCSVä»¥ä¾¿æŸ¥çœ‹
formatted_data.to_csv('formatted_financial_data.csv', index=False)
print("\nâœ… Formatted data saved: formatted_financial_data.csv")

# 4. åˆ›å»ºä»ªè¡¨æ¿
print("\n" + "="*80)
print("ğŸ“Š Visualization 4: Financial Dashboard")
print("="*80)

fig4, axes4 = plt.subplots(2, 2, figsize=(16, 12))
fig4.suptitle('Tesla Financial Dashboard (2020-2024)', fontsize=18, fontweight='bold', y=0.98)

# 4.1 å…³é”®æŒ‡æ ‡å¡ç‰‡
ax1 = axes4[0, 0]
latest_year = '2024'
total_rev_2024 = df_full.loc[df_full['é¡¹ç›®åç§°'] == 'Total revenues', latest_year].values[0]
net_income_2024 = df_full.loc[df_full['é¡¹ç›®åç§°'] == 'Net income', latest_year].values[0]
margin_2024 = (net_income_2024 / total_rev_2024 * 100)

ax1.text(0.5, 0.7, f'{latest_year} Key Metrics', ha='center', va='center', 
         fontsize=14, fontweight='bold', transform=ax1.transAxes)
ax1.text(0.5, 0.5, f'Total Revenue\n${total_rev_2024/1000:.1f}B', ha='center', va='center', 
         fontsize=12, transform=ax1.transAxes)
ax1.text(0.5, 0.3, f'Net Income\n${net_income_2024/1000:.1f}B', ha='center', va='center', 
         fontsize=12, transform=ax1.transAxes)
ax1.text(0.5, 0.1, f'Profit Margin\n{margin_2024:.1f}%', ha='center', va='center', 
         fontsize=12, transform=ax1.transAxes)
ax1.set_xlim(0, 1)
ax1.set_ylim(0, 1)
ax1.axis('off')
ax1.add_patch(plt.Rectangle((0.05, 0.05), 0.9, 0.9, fill=False, 
                           edgecolor='#4ECDC4', linewidth=2))

# 4.2 æ”¶å…¥ç»“æ„æ¼”å˜
ax2 = axes4[0, 1]
categories = ['Automotive', 'Energy', 'Services']
category_colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']

# å‡†å¤‡å †å é¢ç§¯å›¾æ•°æ®
stack_data = []
for category in ['Total automotive revenues', 'Energy generation and storage', 'Services and other']:
    category_values = []
    for year in years:
        value = df_full.loc[df_full['é¡¹ç›®åç§°'] == category, year].values[0]
        category_values.append(value)
    stack_data.append(category_values)

# åˆ›å»ºå †å é¢ç§¯å›¾
ax2.stackplot(years, stack_data, labels=categories, colors=category_colors, alpha=0.7)
ax2.set_xlabel('Year', fontsize=12)
ax2.set_ylabel('Revenue ($M)', fontsize=12)
ax2.set_title('Revenue Structure Evolution', fontsize=14, fontweight='bold')
ax2.legend(loc='upper left', fontsize=10)
ax2.grid(True, alpha=0.3)

# 4.3 åˆ©æ¶¦ç‡å¯¹æ¯”
ax3 = axes4[1, 0]
x_margin = np.arange(len(years))
width = 0.4

ax3.barh(x_margin - width/2, net_income_margins, width, 
         label='Company Margin', color='#FF6B6B', alpha=0.7)
ax3.barh(x_margin + width/2, common_margins, width, 
         label='Common Margin', color='#4ECDC4', alpha=0.7)

ax3.set_yticks(x_margin)
ax3.set_yticklabels(years)
ax3.set_xlabel('Profit Margin (%)', fontsize=12)
ax3.set_title('Margin Comparison', fontsize=14, fontweight='bold')
ax3.legend(fontsize=10)

# 4.4 æ”¶å…¥åˆ©æ¶¦ç›¸å…³æ€§
ax4 = axes4[1, 1]
scatter = ax4.scatter(total_revenues, net_incomes, c=years_num, 
                      cmap='viridis', s=200, alpha=0.7, edgecolors='black')

# æ·»åŠ è¶‹åŠ¿çº¿
z = np.polyfit(total_revenues, net_incomes, 1)
p = np.poly1d(z)
ax4.plot(total_revenues, p(total_revenues), "r--", alpha=0.8, linewidth=2)

ax4.set_xlabel('Total Revenue ($M)', fontsize=12)
ax4.set_ylabel('Net Income ($M)', fontsize=12)
ax4.set_title('Revenue vs Profit Correlation', fontsize=14, fontweight='bold')
ax4.grid(True, alpha=0.3)

# æ·»åŠ å¹´ä»½æ ‡ç­¾
for i, year in enumerate(years):
    ax4.annotate(year, (total_revenues[i], net_incomes[i]), 
                fontsize=9, ha='center', va='bottom')

plt.tight_layout()
plt.savefig('financial_dashboard.png', dpi=150, bbox_inches='tight', facecolor='white')
plt.show()
print("âœ… Financial dashboard saved: financial_dashboard.png")

# 5. åˆ›å»ºçƒ­åŠ›å›¾
print("\n" + "="*80)
print("ğŸ¨ Visualization 5: Financial Heatmap")
print("="*80)

# å‡†å¤‡çƒ­åŠ›å›¾æ•°æ®
heatmap_data = df_full.set_index('é¡¹ç›®åç§°')[years]

# è½¬æ¢ä¸ºåäº¿ç¾å…ƒå•ä½
heatmap_data_gb = heatmap_data / 1000

fig5, ax5 = plt.subplots(figsize=(12, 8))
im = ax5.imshow(heatmap_data_gb, cmap='YlOrRd', aspect='auto')

# è®¾ç½®åˆ»åº¦
ax5.set_xticks(np.arange(len(years)))
ax5.set_yticks(np.arange(len(heatmap_data_gb)))
ax5.set_xticklabels(years)
ax5.set_yticklabels([
    'Auto Sales', 'Reg Credits', 'Auto Leasing', 'Total Auto', 
    'Energy', 'Services', 'Total Revenue', 'Net Income', 
    'Minority Int', 'Common Income'
])

# æ—‹è½¬åˆ»åº¦æ ‡ç­¾
plt.setp(ax5.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# æ·»åŠ æ•°å€¼æ ‡ç­¾
for i in range(len(heatmap_data_gb)):
    for j in range(len(years)):
        value = heatmap_data_gb.iloc[i, j]
        if not pd.isna(value):
            text_color = 'white' if value > heatmap_data_gb.max().max() * 0.6 else 'black'
            ax5.text(j, i, f'{value:.1f}', 
                    ha='center', va='center', 
                    color=text_color, fontsize=9)

ax5.set_title('Tesla Financial Data Heatmap ($B)', fontsize=16, fontweight='bold', pad=20)
plt.colorbar(im, ax=ax5, label='Billions ($)')
plt.tight_layout()
plt.savefig('financial_heatmap.png', dpi=150, bbox_inches='tight', facecolor='white')
plt.show()
print("âœ… Financial heatmap saved: financial_heatmap.png")

print("\n" + "="*80)
print("âœ… Visualization Complete!")
print("="*80)
print("Generated visualization files:")
print("1. ğŸ“Š revenue_composition.png - Revenue composition analysis")
print("2. ğŸ“ˆ financial_trends.png - Revenue & profit trends")
print("3. ğŸ“‹ formatted_financial_data.csv - Formatted data table")
print("4. ğŸ“Š financial_dashboard.png - Financial dashboard")
print("5. ğŸ¨ financial_heatmap.png - Financial data heatmap")
print("\nThese charts provide a comprehensive view of Tesla's financial performance from 2020-2024.")
```
