```python
# Tesla Data Visualization - Complete Code with Path Handling
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from pathlib import Path

# Set Chinese font support
plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial Unicode MS', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# Define file paths - Try multiple possible locations
possible_paths = [
    r"F:\tesla-optimus-business-analysis\data\processed\Tesla_All_Sheets_Scenarios.xlsx",
    r".\data\processed\Tesla_All_Sheets_Scenarios.xlsx",
    r"Tesla_All_Sheets_Scenarios.xlsx",
    r".\Tesla_All_Sheets_Scenarios.xlsx"
]

# Find the file
file_path = None
for path in possible_paths:
    if os.path.exists(path):
        file_path = path
        print(f"Found file at: {path}")
        break

if not file_path:
    print("File not found in any of the expected locations.")
    print("Please check the file exists in one of these paths:")
    for path in possible_paths:
        print(f"  - {path}")
    
    # Ask user for file path
    user_path = input("\nPlease enter the full path to your Excel file: ")
    if os.path.exists(user_path):
        file_path = user_path
    else:
        print("File not found. Exiting...")
        exit()

# Load the Excel file
print(f"\nLoading Excel file: {file_path}")
try:
    xls = pd.ExcelFile(file_path)
    sheets = xls.sheet_names
    print(f"Successfully loaded. Found {len(sheets)} sheets:")
    
    for i, sheet in enumerate(sheets, 1):
        print(f"  {i}. {sheet}")
        
except Exception as e:
    print(f"Error loading file: {e}")
    exit()

# Create a directory for output visualizations
output_dir = "visualizations"
os.makedirs(output_dir, exist_ok=True)
print(f"\nOutput directory created: {output_dir}")

# Function to create visualizations for each sheet
def analyze_sheet(sheet_name, df, sheet_num):
    """Create visualizations for a single sheet"""
    print(f"\n{'='*60}")
    print(f"Analyzing Sheet {sheet_num}: {sheet_name}")
    print(f"{'='*60}")
    print(f"Data shape: {df.shape}")
    print(f"Columns: {list(df.columns)}")
    
    # Identify column types
    numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
    
    print(f"\nNumeric columns: {numeric_cols}")
    print(f"Categorical columns: {categorical_cols}")
    
    # Skip if no numeric columns
    if not numeric_cols:
        print("No numeric columns found. Skipping visualizations.")
        return
    
    # 1. Create summary statistics table
    print("\n1. Summary Statistics:")
    summary_stats = df[numeric_cols].describe().round(2)
    print(summary_stats)
    
    # 2. Create visualization dashboard
    fig = plt.figure(figsize=(16, 12))
    fig.suptitle(f'{sheet_name} - Data Analysis Dashboard', fontsize=16, fontweight='bold')
    
    # Layout: 3x2 grid
    # 1. Top values bar chart
    ax1 = plt.subplot(3, 2, 1)
    if len(numeric_cols) > 0:
        col = numeric_cols[0]
        top_n = min(10, len(df))
        top_data = df.nlargest(top_n, col)[col]
        top_data.plot(kind='bar', ax=ax1, color='steelblue', alpha=0.8)
        ax1.set_title(f'Top {top_n} {col} Values', fontsize=12)
        ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45)
        ax1.grid(True, alpha=0.3)
    
    # 2. Histogram
    ax2 = plt.subplot(3, 2, 2)
    if len(numeric_cols) > 0:
        col = numeric_cols[0]
        df[col].hist(bins=30, ax=ax2, color='lightgreen', alpha=0.7, edgecolor='black')
        ax2.set_title(f'{col} Distribution', fontsize=12)
        ax2.set_xlabel(col)
        ax2.set_ylabel('Frequency')
        ax2.grid(True, alpha=0.3)
    
    # 3. Box plot
    ax3 = plt.subplot(3, 2, 3)
    if len(numeric_cols) > 0:
        df[numeric_cols[0]].plot(kind='box', ax=ax3, vert=False)
        ax3.set_title(f'{numeric_cols[0]} Box Plot', fontsize=12)
        ax3.grid(True, alpha=0.3)
    
    # 4. Line chart (if enough data points)
    ax4 = plt.subplot(3, 2, 4)
    if len(numeric_cols) > 0 and len(df) > 5:
        col = numeric_cols[0]
        n_points = min(50, len(df))
        df[col].head(n_points).plot(kind='line', ax=ax4, marker='o', markersize=3, linewidth=2)
        ax4.set_title(f'{col} Trend (First {n_points} Points)', fontsize=12)
        ax4.set_xlabel('Index')
        ax4.set_ylabel(col)
        ax4.grid(True, alpha=0.3)
    
    # 5. Scatter plot (if at least 2 numeric columns)
    ax5 = plt.subplot(3, 2, 5)
    if len(numeric_cols) >= 2:
        df.plot(kind='scatter', x=numeric_cols[0], y=numeric_cols[1], ax=ax5, alpha=0.6)
        ax5.set_title(f'Scatter: {numeric_cols[0]} vs {numeric_cols[1]}', fontsize=12)
        ax5.grid(True, alpha=0.3)
    
    # 6. Correlation heatmap (if multiple numeric columns)
    ax6 = plt.subplot(3, 2, 6)
    if len(numeric_cols) >= 2:
        corr_matrix = df[numeric_cols[:min(8, len(numeric_cols))]].corr()
        im = ax6.imshow(corr_matrix, cmap='coolwarm', aspect='auto', vmin=-1, vmax=1)
        ax6.set_title('Correlation Heatmap', fontsize=12)
        ax6.set_xticks(range(len(corr_matrix.columns)))
        ax6.set_yticks(range(len(corr_matrix.columns)))
        ax6.set_xticklabels(corr_matrix.columns, rotation=45, ha='right')
        ax6.set_yticklabels(corr_matrix.columns)
        plt.colorbar(im, ax=ax6)
    
    plt.tight_layout()
    
    # Save the figure
    output_path = os.path.join(output_dir, f'{sheet_name.replace("/", "_")}_dashboard.png')
    plt.savefig(output_path, dpi=150, bbox_inches='tight')
    print(f"Dashboard saved to: {output_path}")
    plt.show()
    
    # 3. Create separate charts for each visualization type
    # Pie Chart (if categorical data)
    if categorical_cols:
        cat_col = categorical_cols[0]
        if df[cat_col].nunique() <= 10:
            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
            
            # Pie chart
            value_counts = df[cat_col].value_counts()
            ax1.pie(value_counts.values, labels=value_counts.index, autopct='%1.1f%%')
            ax1.set_title(f'Pie Chart: {cat_col} Distribution', fontsize=14)
            
            # Bar chart
            value_counts.plot(kind='bar', ax=ax2, color='coral', alpha=0.7)
            ax2.set_title(f'Bar Chart: {cat_col} Counts', fontsize=14)
            ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45)
            ax2.set_ylabel('Count')
            ax2.grid(True, alpha=0.3)
            
            plt.suptitle(f'{sheet_name} - Categorical Analysis', fontsize=16)
            plt.tight_layout()
            plt.savefig(os.path.join(output_dir, f'{sheet_name}_categorical.png'), dpi=150)
            plt.show()
    
    # 4. Create line charts for multiple numeric columns
    if len(numeric_cols) > 1:
        fig, ax = plt.subplots(figsize=(12, 6))
        n_points = min(30, len(df))
        
        for col in numeric_cols[:3]:  # Plot first 3 numeric columns
            ax.plot(range(n_points), df[col].head(n_points).values, 
                   marker='o', markersize=3, linewidth=2, label=col)
        
        ax.set_title(f'Line Chart: Multiple Metrics Comparison', fontsize=14)
        ax.set_xlabel('Index')
        ax.set_ylabel('Value')
        ax.legend()
        ax.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.savefig(os.path.join(output_dir, f'{sheet_name}_line_charts.png'), dpi=150)
        plt.show()
    
    print(f"✓ Analysis completed for {sheet_name}")

# Process each sheet
print("\n" + "="*60)
print("STARTING DATA ANALYSIS")
print("="*60)

for i, sheet in enumerate(sheets, 1):
    try:
        print(f"\nProcessing sheet {i}/{len(sheets)}: {sheet}")
        df = pd.read_excel(file_path, sheet_name=sheet)
        analyze_sheet(sheet, df, i)
        
    except Exception as e:
        print(f"Error processing sheet {sheet}: {e}")
        continue

# Create summary report
print("\n" + "="*60)
print("ANALYSIS SUMMARY")
print("="*60)
print(f"Total sheets analyzed: {len(sheets)}")
print(f"Visualizations saved to: {os.path.abspath(output_dir)}")
print("\nSummary of sheets:")

summary_data = []
for i, sheet in enumerate(sheets, 1):
    try:
        df = pd.read_excel(file_path, sheet_name=sheet)
        numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
        categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
        
        summary_data.append({
            'Sheet': sheet,
            'Rows': df.shape[0],
            'Columns': df.shape[1],
            'Numeric Columns': len(numeric_cols),
            'Categorical Columns': len(categorical_cols),
            'Missing Values': df.isnull().sum().sum()
        })
    except:
        summary_data.append({
            'Sheet': sheet,
            'Rows': 'Error',
            'Columns': 'Error',
            'Numeric Columns': 'Error',
            'Categorical Columns': 'Error',
            'Missing Values': 'Error'
        })

summary_df = pd.DataFrame(summary_data)
print("\n" + summary_df.to_string(index=False))

# Save summary to CSV
summary_csv = os.path.join(output_dir, 'analysis_summary.csv')
summary_df.to_csv(summary_csv, index=False, encoding='utf-8-sig')
print(f"\nSummary saved to: {summary_csv}")

print("\n" + "="*60)
print("ANALYSIS COMPLETED SUCCESSFULLY!")
print("="*60)
print(f"\nNext steps:")
print(f"1. Check the '{output_dir}' folder for all visualizations")
print(f"2. Review '{summary_csv}' for detailed statistics")
print(f"3. Use the generated charts for your analysis report")
```
